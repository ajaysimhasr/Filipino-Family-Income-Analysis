# In order to check the corelation between total household income and expenditure features, we plot a heat map of corelation. 
The correlation coefficient measures the strength and direction of a linear relationship between two variables.
observing the plot, we can state that the corelation of different expenses and the total income varies from 0.4 to 0.8
The larger the absolute value of the coefficient, the stronger the relationship between the variables.

```python
cor_matrix = raw_data.corr()
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(cor_matrix, vmax=1, square=True)
plt.show()
```

# In order to select the best feature set, top five features with higher correlation coeficients are selected.
For this a new corelation coefficient map is created

```python
k = 5 
corrmat = raw_data.corr()
columns = cor_matrix.nlargest(k, target)[target].index
First_Five = np.corrcoef(raw_data[columns].values.T)
sns.set(font_scale=1.25)
s = sns.heatmap(First_Five, cbar=True, annot=True, square=True,yticklabels=columns.values, xticklabels=columns.values)
plt.show()
```

```python
foodExpeenseData = pd.DataFrame();
foodExpeenseData['Bread_Cereals_Expenditure']=raw_data['Bread and Cereals Expenditure']
foodExpeenseData['Rice_Expenditure']=raw_data['Total Rice Expenditure']
foodExpeenseData['Meat_Expenditure']=raw_data['Meat Expenditure']
foodExpeenseData['Fish_marine_products_Expenditure']=raw_data['Total Fish and  marine products Expenditure']
foodExpeenseData['Fruit_Expenditure']=raw_data['Fruit Expenditure']
foodExpeenseData['Vegetables_Expenditure']=raw_data['Vegetables Expenditure']
```

```python
foodExpeenseData['Household_Income']=raw_data['Total Household Income']
foodExpeenseData['total_food_expense']=raw_data['Total Food Expenditure']

model_data = pd.DataFrame();
model_data['Household_Income'] = raw_data['Total Household Income']
model_data['food_expenditure'] = raw_data['Total Food Expenditure']
model_data['Housing_water_Expenditure'] = raw_data['Housing and water Expenditure']
model_data['Miscellaneous_Goods_Services_Expenditure']=raw_data['Miscellaneous Goods and Services Expenditure']
model_data['Communication_Expenditure'] = raw_data['Communication Expenditure']
#### do linear regression with food expenditure and include other factors such as number of mobile phones
```

```python
f, ax = plt.subplots(figsize=(20,8))
sns.violinplot(data=foodExpeenseData,bw=5, cut=3, linewidth=1)
ax.set(ylim=(-200000, 1000000))
sns.despine(left=True, bottom=True)
plt.show()
```

```python
import statsmodels.formula.api as smf

multi_linear_model = smf.ols(formula='Household_Income ~ Housing_water_Expenditure+Communication_Expenditure+Miscellaneous_Goods_Services_Expenditure+food_expenditure',data=model_data).fit()
print(multi_linear_model.params)
#print(multi_linear_model.conf_int)
```

```python
print(multi_linear_model.summary())
print("================================================================")
```

```python
import statsmodels.formula.api as smf

linear_model = smf.ols(formula='Household_Income ~ Communication_Expenditure',data=model_data).fit()
print(linear_model.summary())
```

```python
sns.lmplot(x='Communication_Expenditure', y='Household_Income', data=model_data)
sns.lmplot(x='food_expenditure', y='Household_Income', data=model_data)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model


X=raw_data[['Total Food Expenditure','Communication Expenditure']]
y=raw_data['Total Household Income']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


linear_scikit_model = linear_model.LinearRegression()
linear_scikit_model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge

clf = Ridge(alpha=0.1)
clf.fit(X_train, y_train)
```

```python
clf.score(X_train,y_train)
```

# Using cross validation technique to estimate model

```python
predicted_linearModel = cross_val_predict(linear_scikit_model, X_test, y_test, cv=10)
predicted_linearModel
```

```python
fig, ax = plt.subplots()
ax.scatter(predicted_linearModel,y_test, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k-', lw=2)
ax.set_xlabel('Measured')
ax.set_ylabel('predicted_linearModel')
plt.show()
```

```python
predicted_RidgeModel = cross_val_predict(clf, X_test, y_test, cv=10)
predicted_RidgeModel
```

```python
fig, ax = plt.subplots()
ax.scatter(predicted_RidgeModel,y_test, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k-', lw=2)
ax.set_xlabel('Measured')
ax.set_ylabel('predicted_RidgeModel')
plt.show()
```
